```python
# Importo SparkContext
from pyspark import SparkContext
sc = SparkContext("local", "MyProgram")
```

```python
# Leo input y divido en columnas
trips = sc.textFile(my_path + '/Trips/input/')
trips = trips.map(lambda t: t.split("\t"))
```

```python
# Implemente un script de Spark que permita conocer la cantidad de vehículos en movimiento por franja horaria. La duración de la franja horaria es un parámetro de la consulta.
length = sc.broadcast(3000)
```

```python
#tiempo total/length de la franja
def time(timestamp):
    return int(int(timestamp)/length.value) 
```

```python
time_slot = trips.map(lambda t: (time(t[3]), t[0])).distinct() #elimino repetidos
time_slot = time_slot.map(lambda t: (t[0], 1))
time_slot = time_slot.reduceByKey(lambda t1, t2: t1 + t2)
count = time_slot.sortByKey().collect()
for c in count: 
    print('Franja horaria', (c[0] * length.value),'-', ((length.value * (c[0] + 1) - 1)), 'Vehículos en movimiento: ', c[1]) #imprimo franjas y vehículos en movimiento
```
