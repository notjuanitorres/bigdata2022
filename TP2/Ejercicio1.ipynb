```python
from pyspark import SparkContext
sc = SparkContext("local", "MyProgram")
```

```python
# Implemente un script de Spark que permita conocer cuántos viajes realizó cada vehículo. Recordar que un viaje es una serie de coordenadas que finalizan en un destino determinado.
trips = sc.textFile(my_path + '/Trips/input/')
trips = trips.map(lambda t: t.split("\t"))
trips = trips.map(lambda t: (t[0], t[4]))
completedTrips = trips.filter(lambda t: t[1] != '')
completedTripsByID = completedTrips.countByKey()
print('Viajes finalizados por ID: ', completedTripsByID.items())
```
