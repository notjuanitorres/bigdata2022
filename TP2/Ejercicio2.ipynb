```python
# Importo SparkContext
from pyspark import SparkContext
sc = SparkContext("local", "MyProgram")
```

```python
# Implemente un script de Spark que permita conocer cual es el top 3 de los "tipos" de destinos más visitados. Los "tipos" de destino válidos son "Hospital", "Escuela", "Plaza", "Ferretería", "Farmacia", "Supermercado", "Museo", NO interesa contar a los destinos "Otro".
# Leo input y divido en columnas
trips = sc.textFile(my_path + '/Trips/input/')
trips = trips.map(lambda t: t.split("\t"))
```

```python
# Me quedo con destinos, filtro vacíos y "Otro"
destinations = trips.map(lambda t: (t[4], 1))
destinations = destinations.filter(lambda t: t[0] != 'Otro' and t[0] != '')
```

```python
# Aplico reduceByKey y me quedo con el top 3 usando la cantidad como key
destinations = destinations.reduceByKey(lambda t1, t2: t1 + t2)
topDestinations = destinations.top(3, key=lambda t: t[1])
print(topDestinations)
```
