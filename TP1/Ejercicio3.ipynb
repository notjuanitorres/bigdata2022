```python
# Importo emulador 

import os
import sys
sys.path.append('/content/' + root_path)

from MRE import Job
```

```python
# Directorios
outputDir = my_path + 'output/'
inputDir1 = data_path + 'input/Enero/'
inputDir2 = data_path + 'input/Mayo/'
equivalentDir = outputDir + 'equivalent/'
```

```python
def fmap(key, value, context):
    context.write(value, 1)
```

```python
def fmap2(key, value, context):
    context.write(value, 2)
```

```python
def fred(key, values, context):
    sets = set()
    for value in values:
        sets.add(value)
    if len(sets) < 2:
        context.write(key, 1)
```

```python
job = Job(inputDir1, equivalentDir, fmap, fred)
job.addInputPath(inputDir2, fmap2)
success = job.waitForCompletion()
```
